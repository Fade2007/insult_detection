```python
from nltk.tokenize import word_tokenize, sent_tokenize
from gensim.models import Word2Vec
import pandas as pd
```

```python
f = pd.read_csv("../cleaned_dataset.csv")
```

```python
l=[]
l = [[j.lower() for j in word_tokenize(i)] for i in f["comment"]]
#l is a sentence of sentence
print (l)
```

```python
p = Word2Vec(f["comment"])
model = Word2Vec(l,size=150, \
        window=10, \
        min_count=2,\
        workers=10)
model.train(l, total_examples=len(l), epochs=10)
```

```python
w = "fuck"
model.wv.most_similar(w,topn =3) #three top most closest words in the vocabulary.
```
